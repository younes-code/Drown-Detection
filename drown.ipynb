```python
import cv2
import time
import numpy as np
import pygame
import mediapipe as mp
from scipy.spatial import distance as dist
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose
mp_hands = mp.solutions.hands
import os

Drown = False

pTime = 0
cpt = 0
thresh_cons_FRAMES = 100
coronal_right_count=0
coronal_left_count=0
sagital_right_count=0
sagital_left_count=0

coronal=0
sagital=0
# For webcam input:
cap = cv2.VideoCapture(1)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1300)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 750)

with mp_pose.Pose(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as pose:
    while cap.isOpened() and not Drown:
        success, image = cap.read()
        if not success:
            print("Ignoring empty camera frame.")
            # If loading a video, use 'break' instead of 'continue'.
            continue

        # To improve performance, optionally mark the image as not writeable to
        # pass by reference.
        image.flags.writeable = False
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = pose.process(image)

        # Draw the pose annotation on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        mp_drawing.draw_landmarks(
            image,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())
        if results.pose_landmarks:
            keypoints = []
            for data_point in results.pose_landmarks.landmark:
                ih, iw, ic = image.shape
                x,y,z=int(data_point.x*iw),int(data_point.y*ih),int(data_point.z*100)
                keypoints.append([x,y,z])
                #cv2.putText(image,str(data_point.x)+','+str(data_point.y),(x,y),cv2.FONT_HERSHEY_COMPLEX,.8,(0,255,0),2)
                
            coronal_right=dist.euclidean(keypoints[20][1],keypoints[24][1])
            coronal_left=dist.euclidean(keypoints[19][1],keypoints[23][1])
            sagital_right=dist.euclidean(keypoints[20][1],keypoints[24][1])
            sagital_left=dist.euclidean(keypoints[19][1],keypoints[24][1])
            
            if coronal_right >=400 :
                coronal_right_count+=1
                #cv2.putText(image,str(coronal_right_count),(200,200),cv2.FONT_HERSHEY_COMPLEX,.8,(0,255,0),2)
                if coronal_right>thresh_cons_FRAMES:
                    coronal+=1
                    cv2.putText(image, f'sagital and coronal ditected',(100,100),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3)
                    #time.sleep(1)
                    
            else:
                coronal_right_count=0
                
            if coronal_left >=400:
                coronal_left_count+=1
                #cv2.putText(image,str(coronal_left_count),(200,200),cv2.FONT_HERSHEY_COMPLEX,.8,(0,255,0),2)
                if coronal_left>thresh_cons_FRAMES:
                    coronal+=1
                    cv2.putText(image, f'sagital and coronal ditected',(100,100),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3)
                    #time.sleep(1)
            else:
                coronal_left_count=0

                
            if sagital_right >=400 :
                sagital_right_count+=1      
                #cv2.putText(image,str(sagital_right_count),(200,200),cv2.FONT_HERSHEY_COMPLEX,.8,(0,255,0),2)
                if sagital_right>thresh_cons_FRAMES:
                    sagital+=1
                    cv2.putText(image, f'sagital and coronal ditected',(100,100),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3)
                    #time.sleep(1)
            else:
                sagital_right_count=0
                
                
            if sagital_left >=400:
                sagital_left_count+=1
                #cv2.putText(image,str(sagital_left),(200,200),cv2.FONT_HERSHEY_COMPLEX,.8,(0,255,0),2)
                if sagital_left>thresh_cons_FRAMES:
                    sagital+=1
                    cv2.putText(image, f'sagital and coronal ditected',(100,100),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3)
                    #time.sleep(1)
            else:
                sagital_left_count=0
            
    
            
            
        drown=(sagital+coronal)/2
        print(drown)
 

        
        if drown > 50:
            if drown<thresh_cons_FRAMES:
                cv2.putText(image,'DANGER! an alert notification', (100,200),cv2.FONT_HERSHEY_PLAIN,3, (0,0,255),3)
                                    
                                    
                
         
        # Calcul fps
        cTime = time.time()
        fps = 1/(cTime-pTime)
        thresh_cons_FRAMES = fps * 6
        pTime = cTime    
        #cv2.putText(image, f'FPS: {int(fps)}',(20,70),cv2.FONT_HERSHEY_PLAIN,3,(0,255,0),3)
        # Flip the image horizontally for a selfie-view display.
        cv2.imshow('MediaPipe Pose', image)
        cpt += 1
        if cv2.waitKey(1) == 27 :
            break
        
cap.release()
cv2.destroyAllWindows()
```
